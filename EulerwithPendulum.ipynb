```python
# Classwork 02/13 
# angle vs time (This should be a sine wave
# energy vs time
theta = 11*np.pi/180
#y = 100.0
w = 0.0
g = 9.81
L = 1.0
dt = 0.01

# Lists to store the history (for plotting): Initialize these here! 
t_list = []
theta_list = []
w_list = []

# 2. THE SIMULATION LOOP
t = 0
while t < 10: # Run until it hits the ground
    
    # --- TODO: FILL IN THE PHYSICS HERE ---
    # 1. Calculate Acceleration (a = F_net / m)
    # Hint: Remember gravity points down (-), drag opposes velocity
    a = -g/L * theta # <--- REPLACE THIS

    # 2. Update Position (Euler Step)
    theta = theta + w*dt # m <--- REPLACE THIS

    # we fix the issue with simple euler by calculating the speed first and then the angle!
    # 3. Update Velocity (Euler Step)
    w = w + a * dt # m/s <--- REPLACE THIS
    
    # 4. Advance Time
    t += dt
    
    # Store results
    t_list.append(t)
    theta_list.append(theta)
    w_list.append(w)

time_arr = np.array(t_list)
# w_arr = np.array(w_list) I don't think I needed this
thetarr = np.array(theta_list)

# 3. PLOT
#Make sure to add an x label and y label with units, and give it a title and legend. 
plt.figure(figsize=(8,5))
plt.plot(time_arr, thetarr)
plt.xlabel("Time (s)")
plt.ylabel("Angle (m/s)")
plt.title("Simple Euler with Pendulum Angle vs Time")
plt.grid(True)
#plt.legend()
plt.show()

# The issue I was having: I was calling the wrong variable when I wanted to run the while loop
# this caused my plot to not show anything. It took me a while but I finally found the mistake
```

```python
# Energy vs Time: 
# We discussed in class how omega(w) needs to be called before theta to fix the numerical issue
# Corrected angle vs time graph: so that our sine wave isn't increasing in amplitude
# Classwork 02/13 
# angle vs time (This should be a sine wave
# energy vs time
theta = 11*np.pi/180 # converting into radians
#y = 100.0
w = 0.0
g = 9.81
L = 1.0
dt = 0.01

# Lists to store the history (for plotting): Initialize these here! 
t_list = []
theta_list = []
w_list = []
Energy_list = []

# 2. THE SIMULATION LOOP
t = 0
while t < 10: # Run until it hits the ground
    
    # --- TODO: FILL IN THE PHYSICS HERE ---
    # 1. Calculate Acceleration (a = F_net / m)
    # Hint: Remember gravity points down (-), drag opposes velocity
    a = -g/L * theta # <--- REPLACE THIS

    w = w + a * dt 
    theta = theta + w*dt # m <--- REPLACE THIS
    K = 0.5 * L**2 * w**2  # Kinetic energy of the pendulum system
    U = g * L * (1-np.cos(theta)) # Potential energy. 
    E = K + U # total energy. should be constant like we discussed in class
    
    # 4. Advance Time
    t += dt
    
    # Store results
    t_list.append(t)
    theta_list.append(theta)
    Energy_list.append(E)

time_arr = np.array(t_list)
# w_arr = np.array(w_list) I don't think I needed this
thetarr = np.array(theta_list)

# 3. PLOT
#Make sure to add an x label and y label with units, and give it a title and legend. 
plt.figure(figsize=(8,5))
plt.plot(time_arr, Energy_list, color = 'red')
plt.xlabel("Time (s)")
plt.ylabel("Angle (m/s)")
plt.title("Simple Euler with Pendulum Energy vs Time")
plt.grid(True)
#plt.legend()
plt.show()
```
